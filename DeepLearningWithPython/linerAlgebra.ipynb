# Least-Square-Method

```python
 import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def LeastSquareMethod(beProjectedVector,existedVector):
    projectionNum=np.dot(np.linalg.inv(np.dot(beProjectedVector.T,beProjectedVector)),np.dot(beProjectedVector.T,existedVector))
    return np.dot(beProjectedVector,projectionNum)
```

## smallest distance between multi points with Least-Square-Method by projecting to multi dimension

```python
multiPointXWithProToDimensions=np.linspace(-1,1,100)
```

```python
multiPointYWithProToDimensions=((multiPointXWithProToDimensions*multiPointXWithProToDimensions-1)**3+1)*(np.cos(multiPointXWithProToDimensions*2)+0.6*np.sin(multiPointXWithProToDimensions*1.3))
```

```python
offset=multiPointYWithProToDimensions+(np.random.rand(len(multiPointXWithProToDimensions))-0.5)
```

```python
cubeMultiDimensionX=multiPointXWithProToDimensions**3
```

```python
squareMultiDimensionX=multiPointXWithProToDimensions**2
```

```python
expMultiDimensionX=np.exp(multiPointXWithProToDimensions)
```

```python
sinMultiDimensionX=np.sin(multiPointXWithProToDimensions)
```

```python
arrMultiDimensionMatrix=np.vstack((cubeMultiDimensionX,
                                   squareMultiDimensionX,
                                   multiPointXWithProToDimensions,
                                   np.ones(len(multiPointXWithProToDimensions)),
                                   expMultiDimensionX,sinMultiDimensionX)
                                 ).T
```

```python
arrMultiDimensionMatrix
```

```python
multiDimensionOffset=offset.reshape(offset.shape[0],1)
```

```python
multiDimensionOffset
```

```python
multiPointOffsetWithProToDimensions=LeastSquareMethod(arrMultiDimensionMatrix,multiDimensionOffset)
```

```python
multiPointOffsetWithProToDimensions
```

```python
plt.plot(multiPointXWithProToDimensions,multiPointYWithProToDimensions,color='g',linestyle='-',marker='')
plt.plot(multiPointXWithProToDimensions,offset,color='m',linestyle='',marker='o')
plt.plot(multiPointXWithProToDimensions,multiPointOffsetWithProToDimensions,color='r',linestyle='-',marker='')
plt.show()
```
