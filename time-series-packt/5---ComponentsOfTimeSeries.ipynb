```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plot
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 6, 8
plot.style.use('Solarize_Light2')
```

```python
df_vacation = pd.read_csv('Section2/df_vacation.csv',index_col='Month',parse_dates=['Month'])
```

```python
df_vacation.head()
```

```python
# Plot the time series of google searches of the word "vacation"
plot.style.use('bmh')
ax = df_vacation.plot(color='red', grid=True, figsize=(8,5))
ax.set_xlabel('Year')
ax.set_ylabel('Number of Searches')
ax.set_title('Google Trend of Vacation in 2018') 
```

```python
val_column = df_vacation.loc[:, 'Num_Search_Vacation']
```

```python
result = seasonal_decompose(val_column, model='additive', freq=12)
```

```python
type(result)
```

```python
trend_S = pd.Series(result.trend)
```

```python
trend_S.index
```

```python
seasonal_S = pd.Series(result.seasonal)
```

```python
seasonal_S
```

```python
resid_S = pd.Series(result.resid)
```

```python
resid_S
```

```python
observed_S = pd.Series(result.observed)
```

```python
observed_S
```

```python
ts_result = pd.DataFrame(list(zip(trend_S.index, trend_S,seasonal_S,
                        resid_S, observed_S)))
```

```python
ts_result.columns  = ['Month','trend','seasonal', 'resid', 'observed']
```

```python
ts_result.set_index('Month')
```

```python
ts_result.loc[:, 'Month'] = pd.to_datetime(ts_result.loc[:, 'Month'])
```
