```python
import pandas_datareader
import pandas as pd
import matplotlib.pyplot as plot
from pandas_datareader import data
```

```python
from alpha_vantage.timeseries import TimeSeries
```

```python
netflix = data.DataReader("NFLX", start='2009-1-1', end='2019-11-08', data_source='yahoo').loc[:,'Adj Close']
```

```python
type(netflix)
```

```python
pd.DataFrame(netflix).to_csv('netflix.csv')
```

```python
netflix_csv = pd.read_csv('netflix.csv', index_col='Date', parse_dates=['Date'])
```

```python
netflix_csv.head()
```

```python
plot.style.use('fivethirtyeight')
netflix_csv.plot(title='Netflix Adj. Closing Price', 
             figsize=(10,5), color='red')
plot.show()
```

```python
returns = netflix.pct_change()
```

```python
plot.style.use('bmh')
((1 + returns).cumprod() - 1).plot(
      title='Netflix Cumulative Returns', 
      figsize=(10,5), color='purple')
```

```python
returns.plot(figsize = (10, 5), color = 'green')
```
