```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImg(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread('coins.jpg')
```

```python
showImg(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray,'gray')
plt.axis('off')
plt.show()
```

```python
retOne,threshONe=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
```

```python
retTwo,threshTwo=cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
```

```python
retThree,threshThree=cv2.threshold(gray,127,255,cv2.THRESH_TRUNC)
```

```python
retFour,threshFour=cv2.threshold(gray,127,255,cv2.THRESH_TOZERO)
```

```python
retFive,threshFive=cv2.threshold(gray,127,255,cv2.THRESH_TOZERO_INV)
```

```python
images=[gray,threshONe,threshTwo,threshThree,threshFour,threshFive]
```

```python
plt.figure(figsize=(15,5))
```

```python
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.axis('off')
plt.show()
```

```python
showImg(cv2.bitwise_and(image,image,mask=threshONe))
```

```python
licenseImage=imread("license_plate.png")
```

```python
showImg(licenseImage)
```

```python
grayLicenseImage=cv2.cvtColor(licenseImage,cv2.COLOR_RGB2GRAY)
```

```python
medianLicenseImage=cv2.medianBlur(grayLicenseImage,5)
```

```python
retNormal,licenseThresholdingNormal=cv2.threshold(medianLicenseImage,127,255,cv2.THRESH_BINARY)
```

```python
showImg(medianLicenseImage)
```

```python
licenseThresholdingMean=cv2.adaptiveThreshold(medianLicenseImage,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,3)
```

```python
licenseThresholdingGauss=cv2.adaptiveThreshold(medianLicenseImage,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,3)
```

```python
images=[licenseImage,licenseThresholdingNormal,licenseThresholdingMean,licenseThresholdingGauss]
```

```python
plt.figure(figsize=(10,5))
```

```python
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.axis('off')
    
plt.show()
```
