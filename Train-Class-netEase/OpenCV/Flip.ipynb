```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImg(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread("cat.jpg")
```

```python
showImg(image)
```

```python
flipImage=cv2.flip(image,1)
```

```python
showImg(flipImage)
```

```python
flipImageTwo=cv2.flip(image,0)
```

```python
showImg(flipImageTwo)
```

```python
flipImageThree=cv2.flip(image,-1)
```

```python
showImg(flipImageThree)
```

```python
sliceImage=image[0:200,50:300]
```

```python
showImg(sliceImage)
```

```python
rectangle=np.zeros((300,300,3),dtype='uint8')
```

```python
white=(255,255,255)
```

```python
cv2.rectangle(rectangle,(25,25),(275,275),white,-1)
```

```python
showImg(rectangle)
```

```python
circle=np.zeros((300,300,3),dtype='uint8')
```

```python
cv2.circle(circle,(150,150),150,white,-1)
```

```python
showImg(circle)
```

```python
AndImage=cv2.bitwise_and(circle,rectangle)
```

```python
showImg(AndImage)
```

```python
OrImage=cv2.bitwise_or(circle,rectangle)
```

```python
showImg(OrImage)
```

```python
xorImage=cv2.bitwise_xor(circle,rectangle)
```

```python
showImg(xorImage)
```

```python
notImage=cv2.bitwise_not(circle,rectangle)
```

```python
showImg(notImage)
```

```python
kernelOne=cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python
kernelOne
```

```python
kernelTwo=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
```

```python
print(kernelTwo)
```

```python
kernelThree=cv2.getStructuringElement(cv2.MORPH_CROSS,(5,5))
```

```python
kernelThree
```

```python
erosion=cv2.erode(image,kernelOne)
```

```python
showImg(erosion)
```

```python
for i in range(3):
    erosion=cv2.erode(image,kernelOne,iterations=i+1)
    showImg(erosion)
```

```python
dilation=cv2.dilate(image,kernelOne)
```

```python
showImg(dilation)
```

```python
for i in range(3):
    dilation=cv2.dilate(image,kernelOne,iterations=i+1)
    showImg(dilation)
```
