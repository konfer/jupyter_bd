```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImg(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
kernelSizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kernel in enumerate(kernelSizes):
    plt.subplot(1,3,i+1)
    blur=cv2.blur(image,kernel)
    plt.axis("off")
    plt.title("blurred "+str(kernel))
    plt.imshow(blur)
    
plt.show()
```

```python
kernelSizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kernel in enumerate(kernelSizes):
    plt.subplot(1,3,i+1)
    gaussBlur=cv2.GaussianBlur(image,kernel,0)
    plt.axis("off")
    plt.title("gaussBlurred "+str(kernel))
    plt.imshow(gaussBlur)
    
plt.show()
```

```python
kernelSizes=[(3,3),(9,9),(15,15)]
plt.figure(figsize=(15,15))
for i,kernel in enumerate((3,9,15)):
    plt.subplot(1,3,i+1)
    medianBlur=cv2.medianBlur(image,kernel)
    plt.axis("off")
    plt.title("medianBlurred "+str(kernel))
    plt.imshow(medianBlur)
    
plt.show()
```

## Bilateral

```python
params=[(11,21,7),(11,41,21),(15,75,75)]
plt.figure(figsize=(15,15))
for i,(diameter,sigmaColor,sigmaSpace) in enumerate(params):
    plt.subplot(1,3,i+1)
    bilateralBlurred=cv2.bilateralFilter(image,diameter,sigmaColor,sigmaSpace)
    plt.axis("off")
    plt.title("bilateralBlurred "+str((diameter,sigmaColor,sigmaSpace)))
    plt.imshow(bilateralBlurred)
    
plt.show()
```
