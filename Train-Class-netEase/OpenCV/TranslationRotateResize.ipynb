```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def showImg(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread('image.jpg')
```

```python
showImg(image)
```

```python
M=np.float32([[1,0,250],[0,1,500]])
```

```python
M
```

```python
shifted=cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
```

```python
showImg(shifted)
```

```python
(imgH,imgW)=image.shape[:2]
```

```python
imgH
```

```python
imgW
```

```python
(centreX,centreY)=(imgW/2,imgH/2)
```

```python
centreX
```

```python
centreY
```

```python
rotateMatrix=cv2.getRotationMatrix2D((centreX,centreY),45,1.0)
```

```python
rotateImage=cv2.warpAffine(image,rotateMatrix,(imgW,imgH))
```

```python
rotateImage
```

```python
showImg(rotateImage)
```

```python
rotateMatrixTwo=cv2.getRotationMatrix2D((centreX,centreY),60,1.5)
```

```python
rotateImageTwo=cv2.warpAffine(image,rotateMatrixTwo,(imgW,imgH))
```

```python
showImg(rotateImageTwo)
```

```python
resizeImg=cv2.resize(image,(150,150),interpolation=cv2.INTER_NEAREST)
```

```python
showImg(resizeImg)
```

```python
faceImg=imread('face.png')
```

```python
showImg(faceImg)
```

```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
rects=detector.detectMultiScale(faceImg,scaleFactor=1.1,minNeighbors=2,minSize=(10,10),flags=cv2.CASCADE_SCALE_IMAGE)
```

```python
for(x,y,w,h) in rects:
    cv2.rectangle(faceImg,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
showImg(faceImg)
```
