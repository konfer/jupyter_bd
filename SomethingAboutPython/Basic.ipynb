```python
def sqrt_func(x,small):
    assert x>0
    assert small>0
    
    loops =1
    low = 0.0
    high = max(x,1)
    
    while True and loops <=100:
        guess = (low+high)/2
        if abs(guess**2-x)<small:
            break
        elif guess**2>x:
            high=guess
        else:
            low=guess
            
        print("sqrt num： ",x,"|",low,high,guess)
        loops+=1
    
    return guess
```

```python
small_value=0.000001
```

```python
test_data=[10.0,23,25,1,2,25,0.25]
```

```python
for x in test_data:
    y=sqrt_func(x,small_value)
    assert abs(y**2-x)<small_value
    
```
