```r
airline_mean <- dplyr::select(airline, -ID) %>%
group_by(Airline) %>%
summarise_each(funs(mean)) %>%
glimpse()
```

```r
row.names(airline_mean) <- airline_mean$Airline
```

```r
print(airline_mean)
```

```r
airline_mean <- dplyr::select(airline_mean, -Airline)
```
