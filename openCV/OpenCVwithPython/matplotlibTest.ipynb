```python
from PIL import Image
from pylab import *
```

```python
im = array(Image.open('data/empire.jpg'))
```

```python
imshow(im)
x=[100,100,400,400]
y=[200,500,200,500]
plot(x,y,'r*')
plot(x[:2],y[:2])
title('Plotting:"empire.jpg"')
axis('off')
show()
```

```python
imConv = array(Image.open('data/empire.jpg').convert('L'))
figure()
gray()
contour(imConv, origin='image')
axis('off')
contour(imConv, origin='image')
```
