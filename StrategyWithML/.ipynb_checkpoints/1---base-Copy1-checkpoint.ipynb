```r
vars<-c("age","gender","income","house","store_exp","online_exp","store_trans","online_trans")
```

```r
vartype<-c("norm","binom","norm","binom","norm","norm","pois","pois")
```

```r
group_name <- c("Price", "Conspicuous", "Quality", "Style")
```

```r
group_size <- c(250, 200, 200, 350)
```

```r
mus <- matrix(c(

    60, 0.5, 120000,0.9, 500,200,5,2,
    
    40, 0.7, 200000,0.9, 5000,5000,10,10,
    
    36, 0.5, 70000, 0.4, 300, 2000,2,15,
    
    25, 0.2, 90000, 0.2, 200, 2000,2,20),
              
    ncol = length(vars), byrow = TRUE
    
)
```

```r
mus
```

```r
sds<- matrix( 
    
    c(3,NA,8000,NA,100,50,NA,NA,
    
      5,NA,50000,NA,1000,1500,NA,NA,
    
      7,NA,10000,NA,50,200,NA,NA,
    
      2,NA,5000,NA,10,500,NA,NA

     ), 
             
  ncol=length(vars), byrow=TRUE

)
```

```r
isNormCondition <- vartype == "norm"
isPoisCondition <- vartype == "pois"
isBinomCondition <- vartype == "binom"
```

```r
for(i in seq_along(group_name))
{
    cat(i, group_name[i], "\n")
    seg<-data.frame(matrix(NA,nrow=group_size[i], ncol=length(vars))) 
    seg[, isNormCondition] <- rnorm(group_size[i], mean=mus[i,isNormCondition], sd=sds[i,isNormCondition])
    seg[, isPoisCondition] <- rpois(group_size[i], lambda=mus[i,isPoisCondition])
    
    seg[, isBinomCondition] <- rbinom(group_size[i],size=1,prob=mus[i,isBinomCondition])
    
}
     
```
