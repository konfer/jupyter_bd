### By pylab interface

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue')        #  
plt.plot(x, np.sin(x - 1), color='g')           # rgbcmyk 
plt.plot(x, np.sin(x - 2), color='0.75')        # 0~1 
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # RRGGBB00~FF 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB0~1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML
```

```python
plt.plot(x, x + 0, linestyle='solid')        
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')   
plt.plot(x, x + 3, linestyle='dotted');   

#         
plt.plot(x, x + 4, linestyle='-')  #         
plt.plot(x, x + 5, linestyle='--') #         
plt.plot(x, x + 6, linestyle='-.') #         
plt.plot(x, x + 7, linestyle=':');
```

```python
plt.plot(x, x + 0,'-g')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k') #         
plt.plot(x, x + 3, ':r');       

```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x,np.sin(x))
plt.title("A sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),":b",label='cos(x)')
plt.axis('equal')
#plt.legend()
```

### scatter point

```python
scatterX=np.linspace(0,10,30)
scatterY=np.sin(scatterX)
plt.plot(scatterX,scatterY,'o',color='black')
```

```python
rng=np.random.RandomState(0)
```

```python
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:            
    plt.plot(rng.rand(5), rng.rand(5), marker,                     
        label="marker='{0}'".format(marker))            
    plt.legend(numpoints=1)
    plt.xlim(0,1.8)
```
